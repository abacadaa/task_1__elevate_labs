```python
df=df.drop_duplicates()
```

```python
#issue_1-->    Name : whitespace 

print('before correction: ',df["Name"][1])
df["Name"] = df["Name"].str.strip()
df["Name"][1]
```

```python
#issue_2-->    Age : null value, negative value, whitespace

#1  Replace negative ages with NaN
df.loc[df['Age'] < 0, 'Age'] = np.nan  
#2  filling empty spaces
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)  
#3  Convert to integer after handling NaN values
df['Age'] = df['Age'].astype(int)

df['Age']
```

```python
#issue_3-->    Gender : Inconsistent formats (Female, female, FEMALE, F)(Male, male, M)

df['Gender'] = df['Gender'].str.upper()
df['Gender'] = df['Gender'].replace('FEMALE', 'F')
df['Gender'] = df['Gender'].replace('MALE', 'M')
df['Gender'].fillna('unknown')
df['Gender']
```

```python
#issue_4-->    City : Misspelled city

df['City'] = df['City'].str.strip()

# Correcting the misspellings.
city_corrections = {
    'denver': 'Denver', 
    'los angeles': 'Los Angeles', 
    'new york': 'New York',
    'san francisco': 'San Francisco',
    'chicago': 'Chicago',
    'miami': 'Miami',
    'houston': 'Houston',
    'dallas': 'Dallas',
    'seattle': 'Seattle',
    'boston': 'Boston',
    'phoenix': 'Phoenix',
    'austin': 'Austin',
    'portland': 'Portland',
    'san diego': 'San Diego',
    'las vegas': 'Las Vegas',
    'charlotte': 'Charlotte'
}
# 1:convert all to lowercase for matching   2:mapping   3:use title case for presentation.
df['City'] = df['City'].str.lower().map(city_corrections).fillna(df['City'])

df['City']
```

```python
#issue_5-->    Email : Invalid email format (eve@example)

df['Email'] = df['Email'].astype(str)
df['Email'] = df['Email'].str.strip()
def email_is_valid(email):
    # 1. Handle NaN, non-string
        if not isinstance(email, str):
            return False
    # 2. removing whitespace
        email = email.strip()
    # 3. Must contain exactly one '@'
        if email.count('@') != 1:
            return False

        parts = email.split('@')
        local_part = parts[0]
        domain_part = parts[1]

    # 4. Local & Domain part must not be empty
        if not local_part or not domain_part:
            return False
    # 5. Domain part must contain at least one '.'
        if '.' not in domain_part:
            return False
    # 6. Domain part must not start or end with a '.'
        if domain_part.startswith('.') or domain_part.endswith('.'):
            return False
    # 7. Split by '.' and check the last part.
        domain_subparts = domain_part.split('.')            #let domain_subparts be ['example','com']
        if not domain_subparts or not domain_subparts[-1]: # Check if split is empty or last part(from above example 'com') is empty
            return False
        return True

# Create a boolean mask indicating valid emails
is_valid = df['Email'].apply(email_is_valid)

invalid_rows_indices = df[~is_valid].index
df.drop(invalid_rows_indices, inplace=True)

df['Email']
```

```python
#issue_6-->    Phone_Number : Inconsistent formats (123-456-7890,  (987)654-3210, 1112223333, 555.666.7777, 0000000000)


df['Phone_Number'] = df['Phone_Number'].astype(str)
df['Phone_Number'] = df['Phone_Number'].str.strip()

translation_table = str.maketrans("", "", " -().")
def cleaning(n):
    #1 empty values
        if pd.isna(n): 
            return np.nan
            
        cleaned_digits = n.translate(translation_table)
    #2 0000000000
        if cleaned_digits == '0000000000':
            return np.nan
    #3 len
        if len(cleaned_digits) == 10:
           return cleaned_digits
        else:
            return np.nan 

# Apply the cleaning function
df['Phone_Number'] = df['Phone_Number'].apply(cleaning)

df['Phone_Number']
```

```python
#issue_7-->    Subscription_Date    : Inconsistent formats (2023-01-15, 2022/11/01, 05-20-2023)

df['Subscription_Date'] = pd.to_datetime(df['Subscription_Date'], format='mixed')
df['Subscription_Date']
```
